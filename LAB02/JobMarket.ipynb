```python
import pandas as pd
```

```python
#Load the data using Pandas
df = pd.read_csv('job-market.csv')
```

```python
#Visualize top 10 first rows
print(df.head(10))
```

```python
#Fix column datatypes.
df['Company'] = df['Company'].astype('string')
df['Title'] = df['Title'].astype('string')
df['Location'] = df['Location'].astype('string')
df['LowestSalary'] = df['LowestSalary'].astype('float')
df['HighestSalary'] = df['HighestSalary'].astype('float')
df['Requirement'] = df['Requirement'].astype('string')
```

```python
#Check and clean the data
print(df.isnull().sum())

# Drop duplicates
df = df.drop_duplicates()

# Check for duplicates again
print(df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(10, 6))
df.groupby('Title')['HighestSalary'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
f.autofmt_xdate()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 6))
df.groupby('Title')['LowestSalary'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
f.autofmt_xdate()
plt.show()
```
