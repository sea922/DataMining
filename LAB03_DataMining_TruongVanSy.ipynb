```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("job-market.csv")

# Count the number of jobs by location
jobs_by_location = df.groupby("Classification")["Title"].count()

# Sort the locations by number of jobs in descending order
jobs_by_location = jobs_by_location.sort_values(ascending=False)

# Create a bar chart of the jobs by location
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=jobs_by_location.values, y=jobs_by_location.index, palette="RdYlGn")
plt.xlabel("Number of Jobs")
plt.ylabel("Classification")
plt.title("Number of Jobs by Classification")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the data
df = pd.read_csv("job-market.csv")

# Count the number of jobs by location
jobs_by_location = df.groupby("Location")["Title"].count()

# Sort the locations by number of jobs in descending order
jobs_by_location = jobs_by_location.sort_values(ascending=False)

# Count the number of jobs by classification within each location
jobs_by_class = df.groupby(["Location", "Classification"])["Title"].count()

# Define the color scheme for the pie chart
cmap = plt.get_cmap("tab20")
labels = ['Low' , 'Medium' , 'High' , 'Critical']
# Create the pie chart
fig, ax = plt.subplots()
size = 0.3
outer_colors = cmap(np.arange(len(jobs_by_location))*2)
inner_colors = cmap(np.arange(len(jobs_by_class)))
ax.pie(jobs_by_location, radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
ax.pie(jobs_by_class, radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
ax.set(aspect="equal", title='Job Posts by salary range')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("wine.data.csv")

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
numerical_cols = numerical_cols.drop('Label')

corr = df[numerical_cols].corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn_r', ax=ax)

plt.show()
```
